```python
from qiskit_optimization.applications import Maxcut, Tsp
import matplotlib.pyplot as plt
import networkx as nx
import pennylane as qml
from pennylane import numpy as np
from pennylane import qaoa
from functools import partial
```

```python
def draw_graph(G, colors, pos):
    default_axes = plt.axes(frameon=True)
    nx.draw_networkx(G, node_color=colors, node_size=600, alpha=0.8, ax=default_axes, pos=pos)
    edge_labels = nx.get_edge_attributes(G, "weight")
    nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels)
```

```python
# Generating a graph of 3 nodes
n = 4
num_qubits = n**2
tsp = Tsp.create_random_instance(n, seed=123)
adj_matrix = nx.to_numpy_array(tsp.graph)
print("distance\n", adj_matrix)

colors = ["r" for node in tsp.graph.nodes]
pos = [tsp.graph.nodes[node]["pos"] for node in tsp.graph.nodes]
draw_graph(tsp.graph, colors, pos)
```

```python
from itertools import permutations


def brute_force_tsp(w, N):
    a = list(permutations(range(1, N)))
    last_best_distance = 1e10
    for i in a:
        distance = 0
        pre_j = 0
        for j in i:
            distance = distance + w[j, pre_j]
            pre_j = j
        distance = distance + w[pre_j, 0]
        order = (0,) + i
        print(distance, order)
        if distance < last_best_distance:
            best_order = order
            last_best_distance = distance
            print("order = " + str(order) + " Distance = " + str(distance))
    return last_best_distance, best_order


best_distance, best_order = brute_force_tsp(adj_matrix, n)
print(
    "Best order from brute force = "
    + str(best_order)
    + " with total distance = "
    + str(best_distance)
)


def draw_tsp_solution(G, order, colors, pos):
    G2 = nx.DiGraph()
    G2.add_nodes_from(G)
    n = len(order)
    for i in range(n):
        j = (i + 1) % n
        G2.add_edge(order[i], order[j], weight=G[order[i]][order[j]]["weight"])
    default_axes = plt.axes(frameon=True)
    nx.draw_networkx(
        G2, node_color=colors, edge_color="b", node_size=600, alpha=0.8, ax=default_axes, pos=pos
    )
    edge_labels = nx.get_edge_attributes(G2, "weight")
    nx.draw_networkx_edge_labels(G2, pos, font_color="b", edge_labels=edge_labels)


draw_tsp_solution(tsp.graph, best_order, colors, pos)
```

```python
def calculate_path_cost(bitstring, mapping, adj_matrix):
    # Initialize total cost
    total_cost = 0
    # Iterate over the bitstring
    for i, bit in enumerate(bitstring):
        # If the bit is 1, include the edge in the calculation
        if bit == '1':
            edge = mapping[i]
            # Add the weight of the edge to the total cost
            total_cost += adj_matrix[edge[0]][edge[1]]
    return total_cost

# Example usage:

lowest_score = 10e10
lowest_bitstring = ""

for i in range(2**12):
    bitstring = format(i, "012b")
    cost = calculate_path_cost(bitstring, mapping, adj_matrix)
    # discard solution were all nodes are not visited
    if cost < lowest_score and len(bitstring_to_nodes(bitstring, mapping)) == 5:
        print("Bitstring:", bitstring, "Cost:", cost)
        lowest_score = cost
        lowest_bitstring = bitstring


bitstring = most_likely
mapping = {0: (0, 1), 1: (0, 2), 2: (0, 3), 3: (1, 0), 4: (1, 2), 5: (1, 3), 6: (2, 0), 7: (2, 1), 8: (2, 3), 9: (3, 0), 10: (3, 1), 11: (3, 2)}
adj_matrix = nx.to_numpy_array(g)  # Assuming g is your graph defined elsewhere
cost = calculate_path_cost(bitstring, mapping, adj_matrix)
print("Total cost of the path:", cost)
```

```python
def bitstring_to_edges(bitstring, mapping):
    return [mapping[i] for i, bit in enumerate(bitstring) if bit == '1']

def edges_to_path(edges):
    path = []
    # Start from an arbitrary edge
    current_edge = edges[0]
    path.extend(current_edge)
    while len(path) < len(edges) + 1:
        for edge in edges:
            if edge[0] == path[-1] and edge[1] not in path:
                path.append(edge[1])
                break
    return path

# Now let's define a function to calculate the cost of the path.
def calculate_cost_of_path(path, adj_matrix):
    cost = 0
    for i in range(len(path) - 1):
        cost += adj_matrix[path[i], path[i+1]]
    # Adding the cost from last to first to complete the cycle
    cost += adj_matrix[path[-1], path[0]]
    return cost

# Example usage:
bitstring = most_likely
mapping = {0: (0, 1), 1: (0, 2), 2: (0, 3), 3: (1, 0), 4: (1, 2), 5: (1, 3), 6: (2, 0), 7: (2, 1), 8: (2, 3), 9: (3, 0), 10: (3, 1), 11: (3, 2)}
edges = bitstring_to_edges(bitstring, mapping)
path = edges_to_path(edges)
cost = calculate_cost_of_path(path, adj_matrix)

print(f"Path: {path}")
print(f"Cost of the path: {cost}")
```
